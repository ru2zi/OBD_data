```python
# Define a function to get the outlier indices for a given column
def get_outlier_indices(df, col_name):
    Q1 = df[col_name].quantile(0.25)
    Q3 = df[col_name].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_indices = df[(df[col_name] < lower_bound) | (df[col_name] > upper_bound)].index
    return outlier_indices

# Get the outlier indices for the desired columns
columns_to_check = ["TD3", "SC1", "SC2", "SC3"]
outlier_indices_set = set()
for col in columns_to_check:
    outlier_indices = get_outlier_indices(raw_df, col)
    outlier_indices_set.update(outlier_indices)

# Create a new dataframe containing only the outlier rows
raw_df_outliers = raw_df.loc[outlier_indices_set]
raw_df_clean = raw_df.drop(outlier_indices_set)
```

## 변수 생성

```python
#. 주어진 배열 내에서 평균값과 가장 멀리 떨어진 값을 찾습니다.
# 이 함수는 절대값을 사용해 평균과의 차이가 가장 큰 값을 반환합니다.
def find_furthest_from_mean(values):
    mean_value = np.mean(values)
    furthest_value = max(values, key=lambda x: abs(x - mean_value))
    return furthest_value

# 주어진 배열 내에서 평균과의 차이가 가장 작은 값을 찾습니다.
# 이 함수는 제곱을 사용하여 평균과의 차이가 가장 작은 값을 반환합니다.
def find_closest_to_mean(values):
    mean_value = np.mean(values)
    smallest_variance_value = min(values, key=lambda x: (x - mean_value)**2)
    return smallest_variance_value

def get_row_value_with_largest_spread(row, columns):
    values = [row[col] for col in columns]
    return columns[values.index(max(values))]

def get_row_average_for_set(row, column_set):
    selected_value_1 = row[get_row_value_with_largest_spread(row, column_set[0:3])]
    selected_value_2 = row[get_row_value_with_largest_spread(row, column_set[3:6])]
    selected_value_3 = row[get_row_value_with_largest_spread(row, column_set[6:9])]

    return (selected_value_1 + selected_value_2 + selected_value_3) / 3
```

```python
# 튜브 변수와 샤프트 변수 목록
tube_columns = ['TB1', 'TB2', 'TB3', 'TC1', 'TC2', 'TC3', 'TD1', 'TD2', 'TD3']
shaft_columns = ['SA1', 'SA2', 'SA3', 'SB1', 'SB2', 'SB3', 'SC1', 'SC2', 'SC3']
diff_columns = ['SA1_TB1_Diff',	'SA2_TB2_Diff',	'SA3_TB3_Diff','SB1_TC1_Diff',	'SB2_TC2_Diff',	'SB3_TC3_Diff',	'SC1_TD1_Diff','SC2_TD2_Diff','SC3_TD3_Diff']
```

```python
df_clean = raw_df_clean.copy()  # 데이터프레임을 복사하여 새로운 데이터프레임에 할당
```

```python
# 튜브 변수와 샤프트 변수 목록
df_clean['SA1_TB1_Diff'] = df_clean['SA1'] - df_clean['TB1']
df_clean['SA2_TB2_Diff'] = df_clean['SA2'] - df_clean['TB2']
df_clean['SA3_TB3_Diff'] = df_clean['SA3'] - df_clean['TB3']

df_clean['SB1_TC1_Diff'] = df_clean['SB1'] - df_clean['TC1']
df_clean['SB2_TC2_Diff'] = df_clean['SB2'] - df_clean['TC2']
df_clean['SB3_TC3_Diff'] = df_clean['SB3'] - df_clean['TC3']

df_clean['SC1_TD1_Diff'] = df_clean['SC1'] - df_clean['TD1']
df_clean['SC2_TD2_Diff'] = df_clean['SC2'] - df_clean['TD2']
df_clean['SC3_TD3_Diff'] = df_clean['SC3'] - df_clean['TD3']

column_AB = ['SA1_TB1_Diff', 'SA2_TB2_Diff', 'SA3_TB3_Diff']
column_BC = ['SB1_TC1_Diff', 'SB2_TC2_Diff', 'SB3_TC3_Diff']
column_CD = ['SC1_TD1_Diff', 'SC2_TD2_Diff', 'SC3_TD3_Diff']

df_clean['Diff_Furthest_AB'] = df_clean[column_AB].apply(lambda row: find_furthest_from_mean(row), axis=1)
df_clean['Diff_Furthest_BC'] = df_clean[column_BC].apply(lambda row: find_furthest_from_mean(row), axis=1)
df_clean['Diff_Furthest_CD'] = df_clean[column_CD].apply(lambda row: find_furthest_from_mean(row), axis=1)

df_clean['AB_Gap_Value'] = df_clean[column_AB].max(axis=1) - df_clean[column_AB].min(axis=1)
df_clean['BC_Gap_Value'] = df_clean[column_BC].max(axis=1) - df_clean[column_BC].min(axis=1)
df_clean['CD_Gap_Value'] = df_clean[column_CD].max(axis=1) - df_clean[column_CD].min(axis=1)


# Inner Set
df_clean['inner_Average_Selected'] = df_clean.apply(lambda row: get_row_average_for_set(row, ['TB1', 'TB2', 'TB3', 'TC1', 'TC2', 'TC3', 'TD1', 'TD2', 'TD3']), axis=1)

# Outer Set
df_clean['outer_Average_Selected'] = df_clean.apply(lambda row: get_row_average_for_set(row, ['SA1', 'SA2', 'SA3', 'SB1', 'SB2', 'SB3', 'SC1', 'SC2', 'SC3']), axis=1)

# 각 행마다 Tube와 Shaft의 평균에서 가장 멀리 떨어진 값을 찾아 새로운 칼럼에 저장
df_clean['Tube_Furthest_Value'] = df_clean[tube_columns].apply(lambda row: find_furthest_from_mean(row), axis=1)
df_clean['Shaft_Furthest_Value'] = df_clean[shaft_columns].apply(lambda row: find_furthest_from_mean(row), axis=1)
df_clean['Diff_Furthest_Value'] = df_clean[diff_columns].apply(lambda row: find_furthest_from_mean(row), axis=1)

# Calculate the value with the least variance for tube and shaft and save it in new columns
df_clean['Tube_Closest_Value'] = df_clean[tube_columns].apply(lambda row: find_closest_to_mean(row), axis=1)
df_clean['Shaft_Closest_Value'] = df_clean[shaft_columns].apply(lambda row: find_closest_to_mean(row), axis=1)
df_clean['Diff_Closest_Value'] = df_clean[diff_columns].apply(lambda row: find_closest_to_mean(row), axis=1)

# Calculate the max and min values across tube and shaft columns for each row
df_clean['Shaft_Max_Value'] = df_clean[shaft_columns].max(axis=1)
df_clean['Shaft_Min_Value'] = df_clean[shaft_columns].min(axis=1)
df_clean['Tube_Max_Value'] = df_clean[tube_columns].max(axis=1)
df_clean['Tube_Min_Value'] = df_clean[tube_columns].min(axis=1)
df_clean['Diff_Max_Value'] = df_clean[diff_columns].max(axis=1)
df_clean['Diff_Min_Value'] = df_clean[diff_columns].min(axis=1)

# Calculate the combined score for each row
df_clean['Furthest_Distance'] = df_clean['Shaft_Furthest_Value'] - df_clean['Tube_Furthest_Value']
df_clean['Closest_Distance'] = df_clean['Shaft_Closest_Value'] - df_clean['Tube_Closest_Value']
df_clean['Both_big_Distance'] = df_clean['Shaft_Furthest_Value'] - df_clean['Tube_Closest_Value']
df_clean['Both_small_Distance'] = df_clean['Shaft_Closest_Value'] - df_clean['Tube_Furthest_Value']

df_clean['Shaft_max_Tube_min_Distance'] = df_clean['Shaft_Max_Value'] - df_clean['Tube_Min_Value']
df_clean['Shaft_max_Tube_max_Distance'] = df_clean['Shaft_Max_Value'] - df_clean['Tube_Max_Value']
df_clean['Shaft_min_Tube_max_Distance'] = df_clean['Shaft_Min_Value'] - df_clean['Tube_Max_Value']
df_clean['Shaft_min_Tube_min_Distance'] = df_clean['Shaft_Min_Value'] - df_clean['Tube_Min_Value']
```

```python
# 'SA' 세트에서 산포가 가장 큰 값을 가져옴
df_clean['SA_Max_Spread'] = df_clean.apply(lambda row: get_max_spread_value(row, ['SA1', 'SA2', 'SA3']), axis=1)

# 'SB' 세트에서 산포가 가장 큰 값을 가져옴
df_clean['SB_Max_Spread'] = df_clean.apply(lambda row: get_max_spread_value(row, ['SB1', 'SB2', 'SB3']), axis=1)

# 'SC' 세트에서 산포가 가장 큰 값을 가져옴
df_clean['SC_Max_Spread'] = df_clean.apply(lambda row: get_max_spread_value(row, ['SC1', 'SC2', 'SC3']), axis=1)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

def find_furthest_from_mean(values):
    mean_value = np.mean(values)
    furthest_value = max(values, key=lambda x: abs(x - mean_value))
    return furthest_value

def find_closest_to_mean(values):
    mean_value = np.mean(values)
    smallest_variance_value = min(values, key=lambda x: (x - mean_value)**2)
    return smallest_variance_value

def get_row_value_with_largest_spread(row, columns):
    values = [row[col] for col in columns]
    return columns[values.index(max(values))]

def get_row_average_for_set(row, column_set):
    selected_value_1 = row[get_row_value_with_largest_spread(row, column_set[0:3])]
    selected_value_2 = row[get_row_value_with_largest_spread(row, column_set[3:6])]
    selected_value_3 = row[get_row_value_with_largest_spread(row, column_set[6:9])]
    return (selected_value_1 + selected_value_2 + selected_value_3) / 3

def process_and_augment_data(df):
    DF = df.copy()

    # List of tube variables and shaft variables
    tube_columns = ['TB1', 'TB2', 'TB3', 'TC1', 'TC2', 'TC3', 'TD1', 'TD2', 'TD3']
    shaft_columns = ['SA1', 'SA2', 'SA3', 'SB1', 'SB2', 'SB3', 'SC1', 'SC2', 'SC3']
    diff_columns = ['SA1_TB1_Diff', 'SA2_TB2_Diff', 'SA3_TB3_Diff', 'SB1_TC1_Diff', 'SB2_TC2_Diff', 'SB3_TC3_Diff', 'SC1_TD1_Diff', 'SC2_TD2_Diff', 'SC3_TD3_Diff']

    # Calculate the differences
    DF['SA1_TB1_Diff'] = DF['SA1'] - DF['TB1']
    DF['SA2_TB2_Diff'] = DF['SA2'] - DF['TB2']
    DF['SA3_TB3_Diff'] = DF['SA3'] - DF['TB3']

    DF['SB1_TC1_Diff'] = DF['SB1'] - DF['TC1']
    DF['SB2_TC2_Diff'] = DF['SB2'] - DF['TC2']
    DF['SB3_TC3_Diff'] = DF['SB3'] - DF['TC3']

    DF['SC1_TD1_Diff'] = DF['SC1'] - DF['TD1']
    DF['SC2_TD2_Diff'] = DF['SC2'] - DF['TD2']
    DF['SC3_TD3_Diff'] = DF['SC3'] - DF['TD3']

    column_AB = ['SA1_TB1_Diff', 'SA2_TB2_Diff', 'SA3_TB3_Diff']
    column_BC = ['SB1_TC1_Diff', 'SB2_TC2_Diff', 'SB3_TC3_Diff']
    column_CD = ['SC1_TD1_Diff', 'SC2_TD2_Diff', 'SC3_TD3_Diff']

    DF['Diff_Furthest_AB'] = DF[column_AB].apply(lambda row: find_furthest_from_mean(row), axis=1)
    DF['Diff_Furthest_BC'] = DF[column_BC].apply(lambda row: find_furthest_from_mean(row), axis=1)
    DF['Diff_Furthest_CD'] = DF[column_CD].apply(lambda row: find_furthest_from_mean(row), axis=1)

    DF['AB_Gap_Value'] = DF[column_AB].max(axis=1) - DF[column_AB].min(axis=1)
    DF['BC_Gap_Value'] = DF[column_BC].max(axis=1) - DF[column_BC].min(axis=1)
    DF['CD_Gap_Value'] = DF[column_CD].max(axis=1) - DF[column_CD].min(axis=1)

    # Inner Set
    DF['inner_Average_Selected'] = DF.apply(lambda row: get_row_average_for_set(row, ['TB1', 'TB2', 'TB3', 'TC1', 'TC2', 'TC3', 'TD1', 'TD2', 'TD3']), axis=1)

    # Outer Set
    DF['outer_Average_Selected'] = DF.apply(lambda row: get_row_average_for_set(row, ['SA1', 'SA2', 'SA3', 'SB1', 'SB2', 'SB3', 'SC1', 'SC2', 'SC3']), axis=1)

    # For each row, find the value furthest from the average of the tube and shaft and save it in a new column
    DF['Tube_Furthest_Value'] = DF[tube_columns].apply(lambda row: find_furthest_from_mean(row), axis=1)
    DF['Shaft_Furthest_Value'] = DF[shaft_columns].apply(lambda row: find_furthest_from_mean(row), axis=1)
    DF['Diff_Furthest_Value'] = DF[diff_columns].apply(lambda row: find_furthest_from_mean(row), axis=1)

    # Calculate the value with the least variance for tube and shaft and save it in new columns
    DF['Tube_Closest_Value'] = DF[tube_columns].apply(lambda row: find_closest_to_mean(row), axis=1)
    DF['Shaft_Closest_Value'] = DF[shaft_columns].apply(lambda row: find_closest_to_mean(row), axis=1)
    DF['Diff_Closest_Value'] = DF[diff_columns].apply(lambda row: find_closest_to_mean(row), axis=1)

    # Calculate the max and min values across tube and shaft columns for each row
    DF['Shaft_Max_Value'] = DF[shaft_columns].max(axis=1)
    DF['Shaft_Min_Value'] = DF[shaft_columns].min(axis=1)
    DF['Tube_Max_Value'] = DF[tube_columns].max(axis=1)
    DF['Tube_Min_Value'] = DF[tube_columns].min(axis=1)
    DF['Diff_Max_Value'] = DF[diff_columns].max(axis=1)
    DF['Diff_Min_Value'] = DF[diff_columns].min(axis=1)

    # Calculate the combined score for each row
    DF['Furthest_Distance'] = DF['Shaft_Furthest_Value'] - DF['Tube_Furthest_Value']
    DF['Closest_Distance'] = DF['Shaft_Closest_Value'] - DF['Tube_Closest_Value']
    DF['Both_big_Distance'] = DF['Shaft_Furthest_Value'] - DF['Tube_Closest_Value']
    DF['Both_small_Distance'] = DF['Shaft_Closest_Value'] - DF['Tube_Furthest_Value']

    DF['Shaft_max_Tube_min_Distance'] = DF['Shaft_Max_Value'] - DF['Tube_Min_Value']
    DF['Shaft_max_Tube_max_Distance'] = DF['Shaft_Max_Value'] - DF['Tube_Max_Value']
    DF['Shaft_min_Tube_max_Distance'] = DF['Shaft_Min_Value'] - DF['Tube_Max_Value']
    DF['Shaft_min_Tube_min_Distance'] = DF['Shaft_Min_Value'] - DF['Tube_Min_Value']

    # 'SA' 세트에서 산포가 가장 큰 값을 가져옴
    DF['SA_Max_Spread'] = DF.apply(lambda row: get_max_spread_value(row, ['SA1', 'SA2', 'SA3']), axis=1)

     # 'SB' 세트에서 산포가 가장 큰 값을 가져옴
    DF['SB_Max_Spread'] = DF.apply(lambda row: get_max_spread_value(row, ['SB1', 'SB2', 'SB3']), axis=1)

    # 'SC' 세트에서 산포가 가장 큰 값을 가져옴
    DF['SC_Max_Spread'] = DF.apply(lambda row: get_max_spread_value(row, ['SC1', 'SC2', 'SC3']), axis=1)

    df_result = DF[['TB1', 'TC2',  'TD1','SA1',
       'SA2', 'SA3', 'SB1', 'SB2', 'SB3', 'SC1', 'SC2', 'SC3', 'BallSize']]

    # 피처만 추출 (마지막 'BallSize' 제외)
    features = df_result.drop(['BallSize'], axis=1)

    # 데이터 표준화
    scaler = MinMaxScaler()
    scaled_features = scaler.fit_transform(features)
    scaled_features = pd.DataFrame(scaled_features, columns=features.columns)

    # 'BallSize' 컬럼을 결과 데이터프레임에 추가
    df_result = pd.concat([scaled_features, df_result[['BallSize']]], axis=1)

    return df_result

```

```python
def assign_ballsize(value):
    if 4.167 <= value <= 4.171:
        return 4.171
    elif 4.171 < value <= 4.174:
        return 4.172
    elif 4.174 < value <= 4.177:
        return 4.175
    elif 4.177 < value <= 4.179:
        return 4.178
    elif 4.179 < value <= 4.181:
        return 4.180
    elif 4.181 < value <= 4.183:
        return 4.182
    elif 4.183 < value <= 4.185:
        return 4.184
    elif 4.185 <= value <= 4.187:
        return 4.186
    elif 4.187 < value <= 4.189:
        return 4.188
    elif 4.189 < value <= 4.191:
        return 4.190
    elif 4.191 < value <= 4.193:
        return 4.192
    elif 4.193 < value <= 4.195:
        return 4.194
    else:
        return value  # 조건과 맞지 않는 경우 원래의 값을 반환
```

```python
# 'pred_size' 열에 대한 값을 조정하여 'adjusted_size' 열에 할당
test_processed['adjusted_size'] = test_processed['pred_size'].apply(assign_ballsize)

# 결과를 출력
print(test_processed[['pred_size', 'adjusted_size']])
```
